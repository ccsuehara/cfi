```python


replace_dic_se  = {'La mayor parte del tiempo':4, 'Casi siempre':3, 'Algunas veces':2,
       'Casi nunca':1}

replace_dic_uso = {'Entre 2 a 3 horas': 2.5, 'Entre 4 y 5 horas': 4.5, 'Menos de 1 hora': .5,
       'Entre 1 y 2 horas': 1.5, 'De 6 horas a más': 6}

replace_dic_igualdad = {'En desacuerdo': 4, 'De acuerdo': 3, 'Medianamente de acuerdo': 2,
       'Medianamente en desacuerdo': 1}
```

```python
#limpiando un poco la base
df = pd.read_excel('../datos/lb_forms.xlsx', sheet_name = "Sheet1" )
df = df[df['obs_selected'] == 1].copy()

for var in ['se_q1', 'se_q2', 'se_q3', 'se_q4', 'se_q5', 'se_q6',
       'se_q7', 'se_q8', 'se_q9', 'se_q10', 'se_q11', 'se_q12', 'se_q13',
       'se_q14', 'se_q15', 'se_q16', 'se_q17', 'se_q18', 'se_q19', 'se_q20',
       'se_q21', 'se_q22']:
    df.replace({var:replace_dic_se},inplace = True)

df['uso_horas_estimado'] = 0

for var in [ 'uso_altanto', 'uso_redes', 'uso_comunicacion', 'uso_aprender_cole',
       'uso_aprender_propio', 'uso_educ_financiera', 'uso_prevencion',
       'uso_problemas', 'uso_juegos', 'uso_youtube']:
    df[var + "_mod"] =  df[var]
    df.replace({var + "_mod":replace_dic_uso}, inplace = True)
    df.loc[df[var + "_mod"].isna(),var + "_mod" ] = 0
    df['uso_horas_estimado'] += df[var + "_mod"]
    
    
for i in range(1,26):
    df.replace({'igualdad_'+ str(i):replace_dic_igualdad},inplace = True)

df.replace({'igualdad_20':{1:4, 2:3, 3:2,4:1}},inplace = True)

df['index_igualdad'] = df[df.columns[df.columns.str.startswith('igualdad')].tolist()].mean(axis = 1)

#df['uso_reportado'] = df['horas_conectado_clean'] * df['dias_conectado']
```

```python
df['proyeccion_clean'] = df['proyeccion']
```

```python
## Creando las habilidades se
df['toma_decisiones'] = df.eval('(se_q20+se_q21+se_q22+se_q14)/(4)')

df['grit'] = df.eval('(se_q6+se_q7+se_q12)/(3)')
df['escrupulosidad'] = df.eval('(se_q2+se_q11+se_q16)/(3)')
df['amabilidad'] = df.eval('(se_q8+se_q15+se_q18)/(3)')
df['apertura'] = df.eval('(se_q3+se_q10+se_q13)/(3)')
df['extroversion'] = df.eval('(se_q1+se_q4+se_q19)/(3)')
df['estabilidad'] = df.eval('(se_q5+se_q9+se_q17)/(3)')

```
